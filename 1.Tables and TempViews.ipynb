#### Saving to tables

```python
spark.conf.get("spark.sql.warehouse.dir")
```

```python
df2 = df1.where("phone is not null")
df2.show()
```

```python
spark.catalog.listTables()
```

```python
df2.write.format("json").saveAsTable("users_json")
```

```python
df2.write.format("csv").option("header", True).saveAsTable("users_csv")
```

```python
spark.sql("select * from users_json where age > 20").show()
```

```python
spark.sql("select * from users_csv where age > 20 order by age").show()
```

```python
spark.sql("drop table users_csv")
```

```python
ext_path = "E:\\PySpark\\external\\users_ext"
df2.write.format("json").option("path", ext_path).saveAsTable("users_ext")
```

```python
df2.write.format("json").option("path", ext_path).mode("append").saveAsTable("users_ext")
```

```python
spark.sql("select * from users_ext").show()
```

```python
spark.catalog.listTables()
```

```python
spark.sql("drop table users_json")
```
